### Writing files

```python
f = open("demofile.txt", "a")
f.write("Now the file has more content!")
f.close()

#open and read the file after the appending:
f = open("demofile.txt", "r")
print(f.read())
f.close()
```

```python
f = open("demofile2.txt", "w")
f.write("Woops! I have deleted the content!".lower())
f.close()

#open and read the file after the appending:
f = open("demofile2.txt", "r")
print(f.read())
f.close()
```

### Proper way to work with files

```python
with open("exmple.txt", "r") as file:
    txt = file.read()
    
with open("demofile2.txt", "w") as f:
    f.write(txt.lower())

with open("demofile2.txt", "r") as file:
    print(file.read())
```

## Images

```python
import PIL
```

```python
image = '/content/kvadrat63.jpg'
im = PIL.Image.open(image)
display(im)
```

```python
#info
print(im.format, im.size, im.mode)
```

```python
#save
os.makedirs('/content/images/')
im.save('/content/images/original.jpg', quality=95)
```

```python
#resize
width = 180
height = 200
im_resized = im.thumbnail((width, height), Image.ANTIALIAS)
im_resized.save('/content/images/resized.jpg')
display(im_resized)
```

```python
#rotate
angle = 90
im_rot = im.rotate(angle)
im_rot.save(f'/content/images/rot_"{angle}".jpg')
display(im_rot)
```

```python
#transpose
im_flip_lr = im.transpose(Image.FLIP_LEFT_RIGHT)
im_flip_lr.save('/content/images/FLIP_LEFT_RIGHT.jpg')
display(im_flip_lr)

im_flip_tb = im.transpose(Image.FLIP_TOP_BOTTOM)
im_flip_tb.save('/content/images/FLIP_TOP_BOTTOM.jpg')
display(im_flip_tb)

im_flip_rot180 = im.transpose(Image.ROTATE_180)
im_flip_rot180.save('/content/images/ROTATE_180.jpg')
display(im_flip_rot180)
```
