# DuckDuckGo Image Scraper


Create deep learning image datasets with DuckDuckGo Image Scraper library

Based on [joedockrill.github.io/jmd_imagescraper/](https://joedockrill.github.io/jmd_imagescraper/)

Created by [Artem Konevskikh](https://github.com/artem-konevskikh)

```
#@title Install
!pip install -q jmd_imagescraper
from jmd_imagescraper.core import * # dont't worry, it's designed to work with import *

im_size = {
  "Cached": ImgSize.Cached,
  "Small": ImgSize.Small,
  "Medium": ImgSize.Medium,
  "Large": ImgSize.Large,
  "Wallpaper": ImgSize.Wallpaper
}
im_type = {
  "All": ImgType.All,
  "Photo": ImgType.Photo,
  "Clipart": ImgType.Clipart,
  "GIF": ImgType.Gif,
  "Transparent": ImgType.Transparent
}
im_layout = {
  "All": ImgLayout.All,
  "Square": ImgLayout.Square,
  "Tall": ImgLayout.Tall,
  "Wide": ImgLayout.Wide
}
im_color = {
  "All": ImgColor.All,
  "Color": ImgColor.Color,
  "Monochrome": ImgColor.Monochrome,
  "Red": ImgColor.Red,
  "Orange": ImgColor.Orange,
  "Yellow": ImgColor.Yellow,
  "Green": ImgColor.Green,
  "Blue": ImgColor.Blue,
  "Purple": ImgColor.Purple,
  "Pink" : ImgColor.Pink,
  "Brown": ImgColor.Brown,
  "Black" : ImgColor.Black,
  "Gray" : ImgColor.Gray,
  "Teal": ImgColor.Teal,
  "White": ImgColor.White
}

def build_params(max_results, image_size, image_type, image_layout, image_color):
  return {
    "max_results": int(max_results),
    "img_size":    im_size.get(image_size, 'Large'), 
    "img_type":    im_type.get(image_type, 'All'),
    "img_layout":  im_layout.get(image_layout, 'All'),
    "img_color":   im_color.get(image_color, 'All')
  }
```

```
#@title Mount Google Drive
#@markdown Mount Google Drive to save search results.

from google.colab import drive
drive.mount('/content/drive')
```

```
#@title Download images
#@markdown Dataset name. Folder with this name will be created 
dataset_name = 'Mycelium' #@param {type: "string"}
#@markdown Seach query
query = 'mycelium' #@param {type: "string"}
#@markdown Maximum number of results
max_results = 10 #@param {type: "integer"}
#@markdown Image size
image_size = "Large" #@param ["Cached", "Small", "Medium", "Large", "Wallpaper"]
#@markdown Image layout
image_layout = "All" #@param ["All", "Square", "Tall", "Wide"]
#@markdown Image type
image_type = "Photo" #@param ["All", "Photo","Clipart","GIF","Transparent"]
#@markdown Image color
image_color = "All" #@param ["All", "Color", "Monochrome", "Red", "Orange", "Yellow", "Green", "Blue", "Purple", "Pink", "Brown", "Black", "Gray", "Teal", "White"]

params = build_params(max_results, image_size, image_type, image_layout, image_color)

assert dataset_name!='', 'Please, define dataset name'
assert query!='', 'Please, define search query'


root = "/content/drive/MyDrive/scraped-images"
duckduckgo_search(root, dataset_name, query, **params)
```
